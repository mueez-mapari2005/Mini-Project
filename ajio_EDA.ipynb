```python
import pandas as pd # to use pandas

customers = pd.read_csv("customer.csv") # to call csv file in python's dataframe
delivery = pd.read_csv("delivery.csv")
ratings = pd.read_csv("ratings.csv")
products = pd.read_csv("products.csv")
returns = pd.read_csv("returns.csv")
transaction = pd.read_csv("transaction.csv")
orders = pd.read_csv("orders.csv")

orders.describe()
```

```python
# to check for duplicates in your dataframe
orders.duplicated().sum()
transaction.duplicated().sum()
customers.duplicated().sum()

```

```python
# if there are any duplicated values than use
orders = orders.drop_duplicates()
```

```python
# to check for null values in our dataset
orders.isnull().sum()
```

```python
# join tables
customers
```

```python
orders.head()
```

```python
# to join two table together, merge function is used

merged_cust_ord = pd.merge(right = orders, left = customers, how = "inner", on = "C_ID" )
merged_cust_ord
```

```python
# group by is used to summarize data
gb = merged_cust_ord.groupby("State").agg({"Or_ID":"count"})
gb = gb.sort_values(by ="Or_ID", ascending=False)
gb
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize= (12,4))
sns.barplot(x = gb.index, y ="Or_ID", data = gb, hue = gb.index, palette = "rocket")
plt.xticks(rotation = 45)
plt.show()
```
